## <span id="preparandoAmbiente"> 1.0 - Preparando ambiente </span>

Para poder começar a fazer as análises é preciso prepara o ambiente, ou seja, importar as bibliotecas e os dados, e ajustar-los de acordo com necessidades. 

### <span id="importandoBibliotecas"> 1.1 - Importando as bibliotecas </span>

```python
# boas práticas: criar uma célula com todas as bibliotecas necessárias
# importando as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
debug = True
```

### <span id="analisandoGrupo"> 1.2 - Analisando o grupo http://dados.ufrn.br/group/ensino </span>
Esse grupo possui todas as informações da UFRN relacionadas ao ensino de 2009.1 até 2018.1. E dele será pego todas as turmas juntamente com as notas de 2014.1 a 2017.2. 

```python
# Turmas 2014.1, 2014.2, 2015.1, 2015.2, 2016.1, 2016.2, 2017.1, 2017.2
# http://dados.ufrn.br/dataset/turmas
# URLs utiliazadas

# turmas
turmas_url = [
    'http://bit.do/turmas_ufrn_2014_1',
    'http://bit.do/turmas_ufrn_2014_2',
    'http://bit.do/turmas_ufrn_2015_1',
    'http://bit.do/turmas_ufrn_2015_2',
    'http://bit.do/turmas_ufrn_2016_1',
    'http://bit.do/turmas_ufrn_2016_2',
    'http://bit.do/turmas_ufrn_2017_1',
    'http://bit.do/turmas_ufrn_2017_2'
]

# matrículas em componentes (notas)
notas_url = [
    'http://bit.do/notas_ufrn_2014_1',
    'http://bit.do/notas_ufrn_2014_2',
    'http://bit.do/notas_ufrn_2015_1',
    'http://bit.do/notas_ufrn_2015_2',
    'http://bit.do/notas_ufrn_2016_1',
    'http://bit.do/notas_ufrn_2016_2',
    'http://bit.do/notas_ufrn_2017_1',
    'http://bit.do/notas_ufrn_2017_2'
]

# componentes curriculares presenciais
disciplinas_url = 'http://bit.do/componentes_curriculares_ufrn'

if debug:
    turmas_url = ['turmas/' + i.replace('http://bit.do/', '') + '.csv' for i in turmas_url]
    notas_url = ['notas/' + i.replace('http://bit.do/', '') + '.csv' for i in notas_url]
    disciplinas_url = disciplinas_url.replace('http://bit.do/', '') + '.csv'
```

```python
# turmas
semestres = ['2014.1', '2014.2', '2015.1', '2015.2', '2016.1', '2016.2', '2017.1', '2017.2']
turmas = []
for i in range(len(turmas_url)):
    turmas.insert(i, pd.read_csv(turmas_url[i], sep=';'))
    turmas[i]['semestre'] = semestres[i]

# matriculas em componentes
notas = []
for i in range(len(notas_url)):
    notas.insert(i, pd.read_csv(notas_url[i], sep=';'))

# componentes curriculares presenciais
# por alguma razão ou outra, esse dataset tinha algum lixo em linhas que ocasionava 
# erros de leitura, dessa forma foi realizado um filtro apenas por colunas mais importantes
df_disciplinas = pd.read_csv(disciplinas_url, sep=';', usecols=["id_componente","codigo","nivel","nome","unidade_responsavel"])
```

```python
# concatenando todas as listas em um dataframe
df_turmas = pd.concat(turmas)
df_notas = pd.concat(notas)

# limpandos os index após o merge
df_turmas.reset_index(drop=True,inplace=True)
df_notas.reset_index(drop=True,inplace=True)
```

### <span id="reparacaoLimpezaDados"> 1.3 - Reparação e Limpeza dos Dados </span>

Os procedimentos realizados nesta parte do código, tem como objetivo remover todos os dados que estão duplicados que não são uteis para o trabalho. Também é feito o processo para unir as informações de disciplinas e turmas para facilitar o processo de extração dos dados ncessários. 

```python
# Adaptando o dataframe turma e disciplinas para o merge
# Esse procedimento é necessário pois o dataframe turma está codificado com ids

# A coluna id_componente precisa ser a mesma em ambos os dataframes
df_turmas.rename(columns={'id_componente_curricular': 'id_componente'}, inplace=True)

# id_componente em ambos os dataframes precisam ser do mesmo tipo
df_turmas.id_componente = df_turmas.id_componente.astype(str)
df_turmas.id_componente = [i.replace('.0', '') for i in df_turmas.id_componente]

# # Limpando o dataframe disciplinas com linhas desnecessárias
df_disciplinas.dropna(inplace=True)
df_disciplinas = df_disciplinas[~df_disciplinas.id_componente.duplicated()]
```

```python
# Fazendo um merge entre o dataframe turmas e disciplinas utilizando como base a coluna id_componente
# https://pandas.pydata.org/pandas-docs/stable/merging.html
colunas = ["id_turma","id_docente_interno","codigo","nivel_ensino","nome","unidade_responsavel",
          "situacao_turma","capacidade_aluno","descricao_horario", "semestre"]

# merge os dois dataframes utilizando como base a coluna id_componente
merge_turmas_disciplinas = pd.merge(df_turmas, df_disciplinas, on='id_componente')[colunas]

merge_turmas_disciplinas.dropna(inplace=True)
merge_turmas_disciplinas.head()
```

## <span id="filtrandoDados"> 2.0 Filtrando dados relacionados ao IMD </span>

Como a pesquisa é direcionada as disciplinas/turmas do curso do BTI no IMD, é preciso filtrar apenas o necessário, ou seja, apenas os dados relacionadas ao IMD, para então começar a construção dos gráficos.

### <span id="filtrandoTurmas"> 2.1 - Filtrando as turmas </span>

As turmas filtradas foram apenas as que são obrigatórias para o curso do BTI.

```python
# filtrando as turmas com código IMD
turmas_imd = merge_turmas_disciplinas[merge_turmas_disciplinas.codigo.str.contains("IMD")]

# utilizar apenas disciplinas com código "CONSOLIDADA"
turmas_imd = turmas_imd[turmas_imd.situacao_turma == 'CONSOLIDADA']

turmas_imd.capacidade_aluno = turmas_imd.capacidade_aluno.astype(np.int64)
turmas_imd.id_turma = turmas_imd.id_turma.astype(np.int64)
turmas_imd.head()
```

### <span id="filtrandoNotas"> 2.2 - Filtrando as notas </span>

Com a filtragem das turmas, agora é possível fazer a das notas.

```python
# filtrar apenas as turmas do IMD
notas_imd = df_notas[df_notas.id_turma.isin(turmas_imd.id_turma)]

# merge entre os dataframes notas_imd e turmas_imd_nomesprof utilizando como referencia id_turma
notas_imd_merge = pd.merge(notas_imd, turmas_imd, on='id_turma')
# remover alunos duplicados
notas_imd_merge = notas_imd_merge[~notas_imd_merge.duplicated(subset = ['discente', 'id_turma'])]
notas_imd_merge.head()
```

## <span id="criacaoAnaliseGraficos"> 3 - Criação e análise dos gráficos </span>

Os gráficos adotados para expressar como foram as notas de cada disciplina por semestre foi o boxplot ou diagrama de caixa. Ele serve para mostrar a distribuição dos dados, neste caso são as notas. Ele é formado por um primeiro e terceiro quartil, e a mediana ou segundo quartil. Entre o primeiro e terceiro quartil são expressos 50% dos valores, entre o limite inferior até o primeiro são 25% e entre o terceiro e o limite superior os outros 25%. Há também os dados fora da curva, que são valores muito diferentes da maioria dos outros e assim são representados fora dos limites.


![BoxPlot](http://www.portalaction.com.br/sites/default/files/resize/EstatisticaBasica/figuras/boxplot1-700x354.png "Imagem explicativa das divisões de um boxplot")

O tamanho das caixas são proporcionais a dispersão dos dados, ou seja, caso uma caixa seja curta é porque os dados (notas) obtidos foram muito próximos e/ou repetidas, possibilitando mais números fora do curva.

### <span id="disciplinas1"> 3.1 - Disciplinas do 1º Período </span>

O primeiro período possui quatro matérias, sendo três de humanas e uma de matemática. Esse é o semestre com a média final mais alta em todo o curso. 
  * Resolução de Problemas Matemáticos Para TI (RPMTI)
  * Práticas de Leitura e Escrita em Português I (PLE I)
  * Práticas de Leitura em Inglês (PLI)
  * Tecnologia da Informação e Sociedade (TIS)


```python
# carregar a configuração padrão de cores do Seaborn
disciplinas_1 = [
    'RESOLUÇÃO DE PROBLEMAS MATEMÁTICOS PARA TI',
    'TECNOLOGIA DA INFORMAÇÃO E SOCIEDADE',
    'PRÁTICAS DE LEITURA E ESCRITA EM PORTUGUÊS I',
    'PRÁTICAS DE LEITURA E ESCRITA EM INGLÊS'
]
sns.set(font_scale=1.5)

# desenhar um gráfico de barras com a quantidade de estudantes aprovados em disciplinas do 1º período
# quantificados por semestre
for disciplina in disciplinas_1:
    sns.factorplot(x = "semestre", y="media_final",
               data=notas_imd_merge[notas_imd_merge.nome == disciplina],
               kind="box", size=6, aspect=1.5)
    plt.title(disciplina)
    plt.xlabel('Semestre')
    plt.ylabel('Notas')

plt.show()
```

Como era de se esperar, as disciplinas de humanas possuem o primeiro quartil acima de 5 na maioria dos semestres, formando caixas mais curtas. Com exceção de TIS em 2016.2 e PLE I em 2014.2 que apresentaram uma dispersão das notas, possuindo o 1º quartil mais baixo. Já RPMTI a partir de 2016.1 começou a ter notas bastantes concentradas entre 2 e 5, indicando que houve um grande percentual de reprovados. 

### <span id="disciplinas2"> 3.2 - Disciplinas do 2º Período </span>

Um dos períodos mais difícil do curso, pois caso o aluno siga a grade curricular, ele pagará três disciplinas na área de exatas que exigem bastante dos estudantes. Além de ter outras duas de programação e uma de português. Esse é o período com a maior carga horária do curso, então, ele se torna o semestre que mais exige do alunos e isso tem como consequência a essas disciplinas desse período possuírem média final baixas e consequentemente grandes número de reprovados. 

  * Introdução às técnicas de Programação (ITP)
  * Práticas de técnicas de programação (PTP)
  * Cálculo Diferencial e Integral I (Cálculo I)
  * Práticas de Leitura e Escrita em Português II (PLE II)
  * Fundamentos Matemáticos da Computação I (FMC I)
  * Vetores e Geometria Analítica (VGA)


```python
# carregar a configuração padrão de cores do Seaborn
disciplinas_2 = [
    'CÁLCULO DIFERENCIAL E INTEGRAL I',
    'VETORES E GEOMETRIA ANALÍTICA',
    'PRÁTICAS DE LEITURA E ESCRITA EM PORTUGUÊS II',
    'FUNDAMENTOS MATEMÁTICOS DA COMPUTAÇÃO I',
    'INTRODUÇÃO ÀS TÉCNICAS DE PROGRAMAÇÃO',
    'PRÁTICAS DE TÉCNICAS DE PROGRAMAÇÃO'
]
sns.set(font_scale=1.5)

# desenhar um gráfico de barras com a quantidade de estudantes aprovados em disciplinas do 2º período
# quantificados por semestre
for disciplina in disciplinas_2:
    sns.factorplot(x = "semestre", y="media_final",
               data=notas_imd_merge[notas_imd_merge.nome == disciplina],
               kind="box", size=5, aspect=1.5)
    plt.title(disciplina)
    plt.xlabel('Semestre')
    plt.ylabel('Notas')

plt.show()
```

Percebe-se que as matérias de exatas possuem o terceiro quartil menor ou igual a 6 em quase todos os semestres, ou seja, cerca de 75% das notas foram abaixo disso. Já as de programação são muito semelhantes, pois uma é intrinsecamente ligada a outra, sendo que ambas possuem notas bastantes dispersas e não tão altas. Diferentemente PLE II possui notas bastantes semelhantes, entre 5 e 7.

### <span id="disciplinas3"> 3 - Disciplinas do 3º período </span>

O terceiro período possui apenas 4 obrigatórias, sendo duas de matemática e duas de programação.

  * Estrutura de Dados Básicas I (EDB I)
  * Linguagem de Programação I (LP I)
  * Probabilidade
  * Fundamentos Matemáticos da Computação II (FMC II)


```python
# carregar a configuração padrão de cores do Seaborn
disciplinas_3 = [
    'LINGUAGEM DE PROGRAMAÇÃO I',
    'ESTRUTURA DE DADOS BÁSICAS I',
    'PROBABILIDADE',
    'FUNDAMENTOS MATEMÁTICOS DA COMPUTAÇÃO II'
]
sns.set(font_scale=1.5)

# desenhar um gráfico de barras com a quantidade de estudantes aprovados em disciplinas do 3º período
# quantificados por semestre
for disciplina in disciplinas_3:
    sns.factorplot(x = "semestre", y="media_final",
               data=notas_imd_merge[notas_imd_merge.nome == disciplina],
               kind="box", size=5, aspect=1.5)
    plt.title(disciplina)
    plt.xlabel('Semestre')
    plt.ylabel('Notas')

plt.show()
```

As de programação possuem médias um pouco dispersas, variando entre 1.0 e 7.0. Probabilidade as notas são mais concentradas e acima de 5.0 em quase todos os semestres. FMC II sem dúvidas é uma matéria onde os alunos apresentam mais dificuldade, pois nos oito semestres cerca de 75% das notas foram inferiores a 6.0, sendo que em 2017.1 quase que 100% delas foram abaixo de 3, que devido a isso o grande número de alunos que não seguiram esse padrão foi alto, tendo assim alguns pontos foras da curva.

### <span id="disciplinas4"> 3.4 - Disciplinas do 4º período </span>

O quarto período é o último que possui matérias obrigatórias, seguindo a grade curricular de 2014, e são apenas duas de programação.
  * Estrutura de Dados Básicas II (EDB II)
  * Linguagem de Programação II (LP II)


```python
# carregar a configuração padrão de cores do Seaborn
disciplinas_4 = [
    'LINGUAGEM DE PROGRAMAÇÃO II',
    'ESTRUTURAS DE DADOS BÁSICAS II',
]
sns.set(font_scale=1.5)

# desenhar um gráfico de barras com a quantidade de estudantes aprovados em disciplinas do 3º período
# quantificados por semestre
for disciplina in disciplinas_4:
    sns.factorplot(x = "semestre", y="media_final",
               data=notas_imd_merge[notas_imd_merge.nome == disciplina],
               kind="box", size=5, aspect=1.5)
    plt.title(disciplina)
    plt.xlabel('Semestre')
    plt.ylabel('Notas')

plt.show()
```

LP II possi notas mais concentradas entre 4.0 e 7.0 e medianas acima de 5 em todas os semestres, ou seja, há um bom percentual de aprovados.  Já EDB II é um pouco mais disperso, mas que têm medianas acima de 5 em em todos os períodos. Apesar das notas estarem em média 5, o número de aprovados têm desempenho semelhante às disciplinas de exatas. 

## <span id="conclusao"> 4 - Conclusões </span>

É um grande desafio para os professores e coordenação encarar a média final tão baixa em disciplinas da área da matemática e principalmente na área da matemática discreta no curso do BTI.  
O ensino superior está recebendo discentes despreparados para disciplinas de matemática, muito provavelmente devido a deficiência que existe hoje no ensino médio. Isso pode ser reparado com a introdução da disciplina RPMTI no primeiro semestre, com o intuito de revisar todos os conteúdos importantes do ensino médio para as matérias obrigatórias do BTI, e tentar diminuir os grandes números de insucessos nas matérias seguintes. Porém, ela possui médias bastante baixas em vários semestre, o que aparentemente deveria ser mais altas, por se tratar de um “resumo”. Mesmo assim, as outras seguintes continuaram com notas baixas.  
Portanto, fica evidente que medidas devem ser tomadas no ensino médio para que os alunos ingressantes do ensino superior tenham o conhecimento mínimo necessário em disciplinas da área de ciências exatas. Tendo em vista que disciplinas introduzidas no primeiro período com a finalidade de “revisar” o conteúdo, aparentemente não está surtindo grandes efeitos nas matérias seguintes.
